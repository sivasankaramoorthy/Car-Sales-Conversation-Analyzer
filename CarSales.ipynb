```python
%%writefile app.py
import time
import re
import json
import os
import pandas as pd
import streamlit as st
import matplotlib.pyplot as plt
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import fitz  # PyMuPDF

# Download VADER lexicon
nltk.download('vader_lexicon')

DEFAULT_SYSTEM_PROMPT = """\
You are an intelligent assistant tasked with analyzing car sales conversations between customer and salesperson. Your goal is to extract the following details:
1. Customer Requirements for a Car:
  - Car brand name
  - Car Type (Hatchback, SUV, Sedan)
  - Fuel Type
  - Color
  - Distance Travelled
  - Make Year
  - Transmission Type
2. Company Policies Discussed:
  - Free RC Transfer
  - 5-Day Money Back Guarantee
  - Free RSA for One Year
  - Return Policy
3. Customer Objections:
  - Refurbishment Quality
  - Car Issues
  - Price Issues
  - Customer Experience Issues (e.g., long wait time, salesperson behaviour)

Provide the extracted information in JSON format and make sure if a feature is not available in the given input, leave the field empty or as NIL.
"""

B_INST, E_INST = "[INST]", "[/INST]"
B_SYS, E_SYS = "<<SYS>>\n", "\n<</SYS>>\n\n"

def get_prompt(instruction, new_system_prompt=DEFAULT_SYSTEM_PROMPT):
    SYSTEM_PROMPT = B_SYS + new_system_prompt + E_SYS
    prompt_template = B_INST + SYSTEM_PROMPT + instruction + E_INST
    return prompt_template

def generate(text, tokenizer, model):
    try:
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        prompt = get_prompt(text)
        st.write(f"Prompt: {prompt[:500]}...")  # Print a truncated prompt for debugging
        inputs = tokenizer(prompt, return_tensors="pt").to(device)
        start_time = time.time()
        outputs = model.generate(
            **inputs,
            max_new_tokens=128,  # Adjust if needed
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.eos_token_id
        )
        elapsed_time = time.time() - start_time
        st.write(f"Generation Time: {elapsed_time:.2f} seconds")  # Print elapsed time for debugging
        final_outputs = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
        st.write(f"Model Output: {final_outputs[:500]}...")  # Print a truncated output for debugging
        return final_outputs
    except Exception as e:
        st.error(f"Error during text generation: {e}")
        return ""

def extract_information(conversation):
    try:
        structured_response = generate(conversation)

        try:
            json_match = re.search(r'\{.*\}', structured_response, re.DOTALL)
            if json_match:
                json_str = json_match.group(0)
                return json.loads(json_str)  # Parse the JSON string
            else:
                return {"error": "No JSON found in the output"}
        except json.JSONDecodeError as e:
            st.error(f"JSON Parsing Error: {e}")
            return {"error": "Could not parse the JSON output"}
    except Exception as e:
        st.error(f"Error during information extraction: {e}")
        return {"error": "Unexpected error during information extraction"}

def filter_valid_data(data):
    # Filter out NIL or empty values
    return {k: v for k, v in data.items() if v and v != "NIL"}

def create_bar_chart(data, title, xlabel, ylabel):
    if data.empty:
        st.write(f"No data available for {title}")
        return
    fig, ax = plt.subplots()
    data.plot(kind='bar', ax=ax)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    st.pyplot(fig)

def analyze_sentiment(text):
    sia = SentimentIntensityAnalyzer()
    sentiment_score = sia.polarity_scores(text)
    if sentiment_score['compound'] > 0:
        sentiment = 'positive'
    elif sentiment_score['compound'] < 0:
        sentiment = 'negative'
    else:
        sentiment = 'neutral'
    return sentiment, sentiment_score

def aggregate_data(extracted_data_list):
    df = pd.DataFrame(extracted_data_list)

    # Display the DataFrame and its columns for debugging
    st.write("DataFrame Columns:")
    st.write(df.columns)
    st.write("DataFrame Head:")
    st.write(df.head())

    # Filter out invalid or NIL values
    df = df.applymap(lambda x: x if x and x != "NIL" else pd.NA)

    # Aggregating data for visualizations
    color_distribution = df['Car Color'].dropna().value_counts() if 'Car Color' in df.columns else pd.Series()
    price_ranges = df['Price Range'].dropna().value_counts() if 'Price Range' in df.columns else pd.Series()
    car_types = df['Car Type'].dropna().value_counts() if 'Car Type' in df.columns else pd.Series()
    refurbishment_issues = df['Refurbishment Issue'].dropna().value_counts() if 'Refurbishment Issue' in df.columns else pd.Series()
    objections = df['Objection'].dropna().value_counts() if 'Objection' in df.columns else pd.Series()

    return color_distribution, price_ranges, car_types, refurbishment_issues, objections

def extract_text_from_pdf(pdf_file):
    try:
        doc = fitz.open(stream=pdf_file.read(), filetype="pdf")
        text = ""
        for page in doc:
            text += page.get_text()
        return text
    except Exception as e:
        st.error(f"Error reading PDF file: {e}")
        return ""
@st.cache_resource
def load_model():
    model_name = "meta-llama/Llama-2-7b-chat-hf"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        device_map='auto',
        torch_dtype=torch.float16
    )
    return tokenizer, model

tokenizer, model = load_model()
# Streamlit UI
st.title("Car Sales Conversation Analyzer")

st.write("Upload multiple conversation transcripts to extract customer requirements, company policies, and objections, and view visualizations.")

uploaded_files = st.file_uploader("Choose files", type=["txt", "pdf"], accept_multiple_files=True)

if uploaded_files:
    extracted_data_list = []
    sentiment_summary = {"positive": 0, "negative": 0, "neutral": 0}

    for uploaded_file in uploaded_files:
        if uploaded_file.type == "text/plain":
            conversation = uploaded_file.read().decode("utf-8")
        elif uploaded_file.type == "application/pdf":
            conversation = extract_text_from_pdf(uploaded_file)
        else:
            st.error(f"Unsupported file type: {uploaded_file.type}")
            continue

        st.text_area(f"Conversation Transcript: {uploaded_file.name}", conversation, height=300)

        if st.button(f"Extract Information from {uploaded_file.name}"):
            extracted_info = extract_information(conversation)
            time.sleep(20)
            st.json(extracted_info)
            # Filter and append valid data
            valid_data = filter_valid_data(extracted_info)
            extracted_data_list.append(valid_data)

            # Analyze sentiment
            sentiment, sentiment_score = analyze_sentiment(conversation)
            sentiment_summary[sentiment] += 1

            # Provide download option
            json_data = json.dumps(extracted_info, indent=4)
            st.download_button(
                label=f"Download JSON for {uploaded_file.name}",
                data=json_data,
                file_name=f"extracted_info_{uploaded_file.name}.json",
                mime="application/json"
            )

    if extracted_data_list:
        color_distribution, price_ranges, car_types, refurbishment_issues, objections = aggregate_data(extracted_data_list)

        st.subheader("Sentiment Analysis Summary")
        st.write(f"Positive Sentiment Count: {sentiment_summary['positive']}")
        st.write(f"Negative Sentiment Count: {sentiment_summary['negative']}")
        st.write(f"Neutral Sentiment Count: {sentiment_summary['neutral']}")
```
